```python
burglary   = lea.event(0.001)
earthquake = lea.event(0.002)
alarm = lea.joint(burglary,earthquake).switch({ (True ,True ) : lea.event(0.950),
                                                (True ,False) : lea.event(0.940),
                                                (False,True ) : lea.event(0.290),
                                                (False,False) : lea.event(0.001) })
john_calls = alarm.switch({ True  : lea.event(0.90),
                            False : lea.event(0.05) })
mary_calls = alarm.switch({ True  : lea.event(0.70),
                            False : lea.event(0.01) })
```

```python
P(mary_calls.given(alarm))
```

```python
P(mary_calls.given(burglary))
```

```python
P(burglary.given(alarm))
```

```python
P(alarm.given(mary_calls))
```

```python
from lea import *
rain = event(0.20)
sprinkler = if_(rain, event(0.01),
                      event(0.40))
grass_wet = joint(sprinkler,rain).switch({ (False,False): False,
                                           (False,True ): event(0.80),
                                           (True ,False): event(0.90),
                                           (True ,True ): event(0.99)})
```

```python
query = rain.given(grass_wet)
```

```python
print (P(query))
print (P(query.calc()))
print (P(query.calc(algo=EXACT)))
```

```python
%%time
print (P(query.calc(algo=MCRS, nb_samples=10000)))
```

```python
%%time
print (P(query.calc(algo=MCRS, nb_samples=10000, nb_subsamples=10)))
```

```python
print (P(query.calc(algo=MCLW, nb_subsamples=10000)))
```

```python
print (P(grass_wet.given(rain).calc(algo=MCLW, nb_subsamples=10000)))
```
